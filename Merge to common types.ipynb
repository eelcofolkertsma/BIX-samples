This is a python script to merge simple and complex types from xsd into a common library

```python
#Identify folders holding schema-s and output destination

from google.colab import files

uploaded = files.upload()

for fn in uploaded.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(uploaded[fn])))

OUT='IATA_Baggage_CommonTypes.xsd'

XML='<?xml  version="1.0" encoding="UTF-8" ?>'
LIBRARY='<xs:schema targetNamespace="http://www.iata.org/IATA/2015/00/2025.3/IATA_Baggage_CommonTypes" elementFormDefault="qualified" version="1.0" xmlns:xs="http://www.w3.org/2001/XMLSchema" xmlns="http://www.iata.org/IATA/2015/00/2025.3/IATA_Baggage_CommonTypes">'

print('done')
```

```python
import httpx

typesDict={}
count=0
for fn in uploaded.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(uploaded[fn])))
  schema=uploaded[fn].decode('utf-8')
  #print(schema)
  schema=schema.replace("\r\n","")

  n=1
  localDict={}
  while n>0:
        i=schema.find("<xs:simpleType",n)
        item=schema[i : schema.find("</xs:simpleType", i)+16]
        key= 'sT-'+ item[item.find('name="')+6:item.find('">')]
        # item is key value pair where key is simpletype name and value is simpletype object
        #
        typesDict[key]=item
        localDict[key]=item
        n=i+1

  n=1
  while n>0:
        i=schema.find("<xs:complexType name=",n)
        item=schema[i : schema.find("</xs:complexType", i)+17]
        key= 'cT-'+ item[item.find('name="')+6:item.find('">')]
        typesDict[key]=item
        localDict[key]=item
        n=i+1
  count=count+len(localDict)
  print('# of types: '+ str(len(localDict))+'   accumulated: '+ str(count))
  print('consolidated: '+ str(len(typesDict)))


ls=[]
for i in typesDict.items():
    ls.append(i[1])

print(len(ls))
ls.sort()


ls.insert(0, LIBRARY)
ls.insert(0, XML)
ls.append('</xs:schema>')
xsd=''.join(ls)
```

```python
from google.colab import files

with open(OUT, 'w') as f:
  f.write(xsd)

files.download(OUT)
```
