This is a python script to merge simple and complex types from xsd into a common library

```python
#pull in python libraries as required
!pip install xmlschema
import httpx
import xmlschema
```

```python
#Identify registry holding schema-s
PREFIX="https://pinggg.mywire.org/25.3/IATA_Baggage_"
SUFFIX=".xsd"
SCHEMAS=["BagRQ",
         "BagRS"
         ]
```

```python
#Acquire content
schema=httpx.get(PREFIX+SCHEMAS[0]+SUFFIX)
draft = xmlschema.XMLSchema(schema.text)
draft
```

```python
draftlist=[]
for xsd_component in draft.iter_components():
    #print(str(xsd_component))
    draftlist.append(str(xsd_component))
dl=sorted(draftlist)
dl
```
