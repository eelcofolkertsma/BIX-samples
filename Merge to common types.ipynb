This is a python script to validate your json documents files to BIX v1 25.3 schema

```python
#pull in python libraries as required
import httpx
import json
from referencing import Registry, Resource
from referencing.exceptions import NoSuchResource
from jsonschema import Draft202012Validator
```

```python
#Identify registry holding library, schema and document
REGISTRY = "https://pinggg.mywire.org/json/"#web server with schemas
source=REGISTRY #schema host
LIBRARY=REGISTRY+"demo-things-library-a1.v0.1.0.json"#$id of library
SCHEMA=REGISTRY+"demo-thing-instruction-a1.v0.1.0.json"#$id of relevant schema
DOCUMENT=REGISTRY+"thing-instruction-document.json"#local id of message
print(DOCUMENT)
```

```python
#Acquire content
document = httpx.get(DOCUMENT).json() #pull in the message
contents = httpx.get(LIBRARY).json()
resource=Resource.from_contents(contents)
initial_registry=Registry()
registry = resource @ initial_registry
print(registry)
```

```python
validator = Draft202012Validator(
    httpx.get(SCHEMA).json(), #prepare validator for schema
        registry=registry,  # to resolve $refs
)

print(document)
validator.validate(document) # actual validation
```
