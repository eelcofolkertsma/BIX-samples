This is a python script to merge simple and complex types from xsd into a common library

```python
#Identify registry holding schema-s and output destination
OUT='IATA_Baggage_CommonTypes.xsd'

PREFIX="https://pinggg.mywire.org/draft/IATA_Baggage_"
SUFFIX=".xsd"
SCHEMAS=["BagRQ",
         "BagRS",
         "BagNotifRQ",
         "CartRQ",
         "CartRS",
         "CartNotifRQ",
         "ULD_RQ",
         "ULD_RS",
         "ULD_NotifRQ",
         ]
XML='<?xml  version="1.0" encoding="UTF-8" ?>'
#LIBRARY='<xs:schema targetNamespace="http://www.iata.org/IATA/2015/00/2025.3/IATA_Baggage_CommonTypes" xmlns:dsig="http://www.w3.org/2000/09/xmldsig#" elementFormDefault="qualified" version="1.0" xmlns:xs="http://www.w3.org/2001/XMLSchema" xmlns="http://www.iata.org/IATA/2015/00/2025.3/IATA_Baggage_CommonTypes">'
LIBRARY='<xs:schema targetNamespace="http://www.iata.org/IATA/2015/00/2025.3/IATA_Baggage_CommonTypes" elementFormDefault="qualified" version="1.0" xmlns:xs="http://www.w3.org/2001/XMLSchema" xmlns="http://www.iata.org/IATA/2015/00/2025.3/IATA_Baggage_CommonTypes">'
#DSIG='<xs:import namespace="http://www.w3.org/2000/09/xmldsig#" schemaLocation="xmldsig-core-schema.xsd"/> '
```

```python
import httpx

typesList=[]
for s in SCHEMAS:
    print(s)
    schema=httpx.get(PREFIX+s+SUFFIX).text
    # string.find(substring, start, end)
    # string[start, end]
    # list.insert(pos, elmnt)
    # string.replace(oldvalue, newvalue, count)

    schema=schema.replace("\r\n","")

    n=1
    while n>0:
        i=schema.find("<xs:simpleType",n)
        typesList.append(schema[i : schema.find("</xs:simpleType", i)+16])
        n=i+1

    n=1
    while n>0:
        i=schema.find("<xs:complexType name=",n)
        # add filter to prevent three almost duplicates for pack-into
        typesList.append(schema[i : schema.find("</xs:complexType", i)+17])
        n=i+1
    print(len(typesList))

typesList.sort()
typesList = list(dict.fromkeys(typesList))
print(len(typesList))
#typesList.insert(0, DSIG)
typesList.insert(0, LIBRARY)
typesList.insert(0, XML)
typesList.append('</xs:schema>')
xsd=''.join(typesList)
```

```python
from google.colab import files

with open(OUT, 'w') as f:
  f.write(xsd)

files.download(OUT)
```
