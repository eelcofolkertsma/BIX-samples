This is a python script to collect dependencies between types in xml library

```python
#Identify folders holding schema-s and output destination

from google.colab import files

uploaded = files.upload()

for fn in uploaded.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(uploaded[fn])))

OUT='IATA_Baggage_CommonTypes.xsd'

print('done')
```

```python
# create dependency map and select roots
types=[]
for fn in uploaded.keys():
  print(fn)
  schema=uploaded[fn].decode('utf-8')
  #print(schema)
  schema=schema.replace("\r\n","")

n=1
while n>0:
    i=schema.find("<xs:complexType name=",n)
    item=schema[i : schema.find("</xs:complexType", i)+17]
    parent= item[item.find('name="')+6:item.find('">')]
    m=1
    while m>0:
        j=item.find('type="',m)#perhaps refine and only include child of complexType
        child=item[j+6:item.find('"', j+7)]
        types.append([parent,child])
        if item.find('type="',j+1)>-1:
            m=j+1
        else:
            m=0
    if schema.find("<xs:complexType name=",i+1)>-1:
        n=i+1
    else:
        n=0

print('done')
parents=[]
childs=[]
for i in types:
    parents.append(i[0])
    childs.append(i[1])

roots=[]
for k in parents:
    if k not in childs:
        roots.append(k)
print(list(dict.fromkeys(roots)))
```
